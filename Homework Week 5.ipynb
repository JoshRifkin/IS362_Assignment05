```python
import pandas as pd
```

```python
airports = "airports.csv"
weather = "weather.csv"
ap_data = pd.read_csv(airports)
wthr_data = pd.read_csv(weather)
```

<h3>Finding the Northernmost airport in the United States</h3>
<p> Finding the northernmost airport is not a terribly difficult task. It can be completed simply by sorting the airports in our dataframe by the latitudinal coordinates of each airport. Latitude is the measurment for angular distance to the north and south of the equator. The equator is located at 0 Degrees, the north pole at 90 degrees north, and the south pole at 90 degrees south.</p>

```python
# Sort the values by the highest latitude, then print only the top 5 slots.
ap_data.sort_values(by=['lat'], ascending = False).head()
```

<p>As you can see above, we printed a list of the five northernmost airports. Doing some research shows the the number one entry above seems to be a mistake. The latitude and longitude are switched.</p>

<h3>Finding the Easternmost airport in the United States</h3>
<p> This can be done in much the same way as the previous problem but instead of sorting by the latitude, we will sort by longitude. Longitude is the measurment for angular distance to the east and west of the meridian in Greenwich, England.</p>

```python
# Sort the values by the highest longitude, then print only the top 5 slots.
ap_data.sort_values(by=['lon'], ascending = True).head()
```

<p>In the table above, we see that ADK is labeled as the easternmost airport in the United States, however, if you look at a map, you can clearly see that this airport is very far west of the United States. The reason for this is that the airport in fact crosses over the meridian line, crossing from being technically to the "west" to being the most "east".<p>


<p>To find the most eastern airport as we would otherwise imagine it, we will limit our table to the "New York" timezone which includes the entire eastern seaboard of the United states.</p>

```python
ap_data[ap_data['tzone'] == 'America/New_York'].sort_values(by=['lon'], ascending = False).head()
```

<p>Now, as you can see above, we've determined that the "easternmost" airport is EPM - Eastport Municipal Airport located in Maine.</p>

<h3>Finding the Windiest airport in the United States on February 12, 2013</h3>


```python
wthr_data = wthr_data[(wthr_data['month'] == 2) & (wthr_data['day'] == 12) & (wthr_data['year'] == 2013)]
```

```python
wthr_data.sort_values(by=['wind_speed'], ascending=False).head()
```

<p>As we can see in the chart above, the - seemingly - windiest airport on February 12, 2013 was EWR with wind speeds of 1048 MPH. That seemed rather high so I performed a quick google search for the windiest day in the United States. As we can see from the screenshot below. The fastest wind speeds on record took place on April 12, 1934 in Mount Washington, NH with a max wind speed of 231 MPH. As such we can determine that the first line is an error. Therefore the second entry is the correct answer. As such LGA was the windiest airport on February 12, 2013 with windspeeds of up to approximately 23 MPH.</p>

<p>We might also consider the fact that perhaps we can measure the windiest airport based on average windspeed. However, the mistaken entry seen above will greatly swing the average. As such, we must remove that entry prior to taking the average.</p>

```python
wthr_data.drop([1009], inplace = True)
wthr_data.reset_index(inplace = True)
```

```python
wthr_data.sort_values(by=['wind_speed'], ascending=False).head()
```

```python
wthr_data.groupby('origin').mean().sort_values(by=['wind_speed'], ascending=False).head()
```

<p>As you can see, the windiest airport still remains LGA with an average windspeed of 14.96 MPH on February 12, 2013.</p>
